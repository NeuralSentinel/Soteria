```python
import json
import os
import pandas as pd
```

```python
lang_high = ['English', 'Spanish', 'German', 'French', 'Chinese']
lang_mid = ['Arabic', 'Bulgarian', 'Thai', 'Hindi']
lang_low = ['Bengali', 'Tamil', 'Telugu']
```

```python
models = ['llama3_1', 'Mistral', 'Phi', 'Qwen']
```

```python
for model in models:
    dicti = {}
    for lang in lang_low:
        head_file = os.path.join(f'{model}_5_Dataset',f'{lang}_Heads.json')
        with open(head_file, 'r') as file:
            data = json.load(file)
        for val in data:
            val1 = tuple(val)
            if val1 not in dicti.keys():
                dicti[val1] = 1
            else:
                dicti[val1] += 1
    val_based_rev = {k: v for k, v in sorted(dicti.items(), key=lambda item: item[1], reverse=True)}
    count = [0,0,0,0,0]
    for key, item in val_based_rev.items():
        count[item-1] += 1
    print(count)
    new_dict = {}
    for key, value in val_based_rev.items():
        if value >= 2:
            new_dict[key] = value
    print(len(new_dict))
    save_file = os.path.join(f'{model}_5_Dataset','Low_Lang.json')
    with open(save_file, 'w',encoding='utf-8') as f:
        json.dump(list(new_dict), f)
    # break
```
