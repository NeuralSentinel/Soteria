```python
import torch
```

```python
datasets = ['ag_news','antonym', 'country-capital', 'country-currency', 'sentiment']
```

```python
# list_of_set_top_20=[]
avg_effect_head = {}
for i in range(32):
    for j in range(32):
        avg_effect_head[tuple([i,j])] = 0
for data in datasets:
    ide_array=torch.load(f'../results/Llama3_1_{data}/{data}_Tamil/{data}_Tamil_indirect_effect.pt', weights_only=True)
    print(ide_array)
    for i, trial in enumerate( range(0,25)):
        print(i)
        ide_array_2d=ide_array[trial]
        coordinates = [(i, j) for i in range(ide_array_2d.shape[0]) for j in range(ide_array_2d.shape[1])]
        values = [ide_array_2d[i, j] for i, j in coordinates]
        value_coords = list(zip(values, coordinates))
        # print(len(value_coords))
        # break
        sorted_value_coords = sorted(value_coords, key=lambda x: x[0], reverse=True)
        # print(len(sorted_value_coords))
        # print(sorted_value_coords)
        sorted_dict = {coord: value for value, coord in sorted_value_coords}
        for coord, value in sorted_dict.items():
            avg_effect_head[tuple(coord)] += value.item()
        # print(len(sorted_dict))
        # break
        # print(sorted_dict)
        # list_of_coordinates=list(sorted_dict.keys())
        # print(len(list_of_coordinates))
        # break
        # list_of_set_top_20.append(list_of_coordinates[0:1024])
```

```python
# avg_effect_head
```

```python
for item, val in avg_effect_head.items():
    avg_effect_head[item] = val/(25*5)
```
